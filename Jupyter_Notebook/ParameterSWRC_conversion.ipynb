## Definizione dei parametri

```python
thetaS=0.3
thetaR=0.05
n=1.31
alpha=6
m=1-(1/n)
```

## Calcolo suzione con inversione curva di ritenzione VG

```python
def psi(thetaS,thetaR,n,alpha,m,theta):
    psi_0=(1/alpha**(n))*(((theta-thetaR)/(thetaS-thetaR))**(-1/m)-1)
    psi=psi_0**(1/n)
    return psi
```

```python
psi25=psi(thetaS,thetaR,n,alpha,m,vwcMean25)
psi25
```

```python
psi75=psi(thetaS,thetaR,n,alpha,m,vwcMean75)
psi75
```

```python
psi125=psi(thetaS,thetaR,n,alpha,m,vwcMean125)
psi125
```

```python
psi175=psi(thetaS,thetaR,n,alpha,m,vwcMean175)
psi175
```

```python
def thetawp(ts,tr,psi,n,alpha):
    m=1-(1/n)
    thetawp=(ts-tr)*((tr/(ts-tr))+(1+abs(alpha*psi)**n)**(-m))
    return thetawp
```

```python
thetawp(thetaS,thetaR,-psi125,n,alpha)
```
